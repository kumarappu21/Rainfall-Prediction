```python
df=df1.iloc[list(df1['STATE_UT_NAME'] == 'HARYANA')]
df
```

```python
w = sns.barplot(x='DISTRICT', y='ANNUAL', data=df, order=df.sort_values('ANNUAL', ascending=False).DISTRICT, palette='magma')
w.set_xticklabels(w.get_xticklabels(), rotation= 40, ha='right')
sns.set(rc={'figure.figsize':(20,8)})
```

```python
from sklearn.model_selection import train_test_split
x = df[['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']]
y= df.ANNUAL
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
lr.coef_, lr.intercept_
```

```python
lr.predict([[10,20,46,86]])
```

```python
y_pred=lr.predict(x_test)
```

```python
plt.scatter(y_pred, y_test, color='g')
plt.plot(y_pred, y_test, color='r')
```
